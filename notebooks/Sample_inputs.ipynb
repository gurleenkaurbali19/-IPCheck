```python
import pandas as pd
import random
```

```python
df=pd.read_csv('Log_dataset_combined.csv')
df.head()
```

```python
SAMPLE_IP_COUNT = 25   
SAMPLE_ROWS_PER_IP_MIN = 5
SAMPLE_ROWS_PER_IP_MAX = 15
SAMPLE_FILES = 2    
```

```python
eligible_ips = []
ip_counts = df['IP_Address'].value_counts() 
```

```python
for ip, count in ip_counts.items():
    if count >= SAMPLE_ROWS_PER_IP_MIN:
        eligible_ips.append(ip)
```

```python
eligible_ips
```

```python
for file_n in range(SAMPLE_FILES):
    sampled_ips = random.sample(eligible_ips, SAMPLE_IP_COUNT)
    sample_rows = []

    for ip in sampled_ips:
        ip_rows = df[df["IP_Address"] == ip]
        n_rows = min(random.randint(SAMPLE_ROWS_PER_IP_MIN, SAMPLE_ROWS_PER_IP_MAX), len(ip_rows))
        # Sample WITHOUT replacement, so no duplicate rows for this IP in the same file
        ip_sample = ip_rows.sample(n=n_rows, replace=False, random_state=random.randint(0, 99999))
        sample_rows.append(ip_sample)


    final_df = pd.concat(sample_rows).sample(frac=1, random_state=random.randint(0, 99999)).reset_index(drop=True)


    # Saving to CSV
    final_df.to_csv(f"sample_test_log_{file_n+1}.csv", index=False)

print("Sample CSV files created successfully!")
```
